### **Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.**

-> Linear Regression is used to predict a continuous value.

Example:



*   Predict house price
*   Predict sales

-> Linear Regression Equation

y = b0 + b1x

Where:



*   y = Target variable
*   x = Feature


*   b0 = Intercept
*   b1 = Coefficient


-> Intercept (b)

The intercept is the value of y when x = 0.

It represents the starting point of the regression line on the y-axis.

-> Coefficient (m)

The coefficient tells us how much the output changes when the input increases by one unit.





### **Q4. Explain the following evaluation metrics used in Linear Regression:**



### *   **SSE**
### *   **MSE**


### *   **RMSE**
### *   **R² Score**


### *   **Adjusted R²**







-> **SSE (Sum of Squared Errors)**

SSE measures the total squared error between actual values and predicted values.

Formula : SSE = Σ (y - ŷ)²

Where:



*   y = Actual value
*   ŷ = Predicted value

Lower SSE means the model is better.


-> **MSE (Mean Squared Error)**

MSE is the average of squared errors.

Formula : MSE = SSE / n

Where:



*   SSE = Sum of Squared Errors
*   n = Number of observations

MSE tells us how far predictions are from actual values on average. Lower MSE means better model.

-> **RMSE (Root Mean Squared Error)**

RMSE is the square root of MSE

Formula : RMSE = √MSE

RMSE is useful because it is in the same units as the target variable. Lower RMSE means better model.

-> **R-Squared (R²)**

R² measures how well the model explains the variation in the data.

Formula : R² = 1 - (SSE / SST)

Where:



*   SSE = Sum of Squared Errors
*   SST = Total Sum of Squares

R² Value Range:

0 → Model explains nothing 1 → Model explains everything

Higher R² = Better model

-> **Adjusted R-Squared**

Adjusted R² is a modified version of R² that adjusts for the number of features.

R² always increases when we add more features, even if they are useless. Adjusted R² increases only when the new feature improves the model.

Formula : Adjusted R² = 1 - [(1 - R²)(n - 1) / (n - p - 1)]

Where :



*   n = Number of observations
*   p = Number of features

### **Q10. Train a Linear Regression model and perform the following:**


### *   **Make predictions**
### *  **Calculate MAE**


### *   **Calculate MSE**
### *   **Calculate RMSE**


### *   **Calculate R² Score**
### *   **Calculate Adjusted R²**


### *   **Write the Linear Regression equation**
### *   **Interpret the model coefficients**









```python
# Train the Model

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Load dataset
df = pd.read_csv("BostonHousing.csv")

# Features and Target
X = df.drop('medv', axis=1)
y = df['medv']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

# Train Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make Predictions

y_pred = model.predict(X_test)

print(y_pred[:10])
```

```python
# Calculate MAE, MSE, RMSE, and R²

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python
# Calculate Adjusted R²

n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))

print("Adjusted R²:", adjusted_r2)
```

```python
# Write the Linear Regression Equation

# Intercept:

print("Intercept:", model.intercept_)
```

```python
# Coefficients:

coeff_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print(coeff_df)
```

```python
# Using the fitted model:

print(model.intercept_)
print(model.coef_)
```

```python
# Interpret the Model Coefficients

coeff_df.sort_values(by='Coefficient', ascending=False)
```

### **Final Interpretation**



*   The Linear Regression model predicts house prices (medv) using 13 housing-related features.
*   The model explains approximately 67% of the variation in house prices (R² ≈ 0.67).



*   RMSE ≈ 4.93, meaning predictions are off by about 4.93 price units on average.
*   rm (average rooms) generally has a strong positive effect on house prices.


*   lstat (lower-status population percentage) typically has the strongest negative effect on house prices.
*   The Adjusted R² remains close to the R² value, indicating that most features contribute useful information to the model.
